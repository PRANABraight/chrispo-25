```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Set random seed for reproducibility
np.random.seed(42)

# Define constants
PARTICIPANTS = 300
DAYS = 5
SPORTS = ['Basketball', 'Swimming', 'Tennis', 'Athletics', 'Volleyball',
          'Soccer', 'Boxing', 'Cycling', 'Gymnastics', 'Martial Arts']

# Generate base data
data = {
    'Participant_ID': range(1, PARTICIPANTS + 1),
    'Age': np.random.randint(18, 45, PARTICIPANTS),
    'Gender': np.random.choice(['M', 'F'], PARTICIPANTS),
    'Sport_Event': np.random.choice(SPORTS, PARTICIPANTS),
    'Experience_Years': np.random.randint(1, 15, PARTICIPANTS),
    'Performance_Score': np.random.uniform(60, 100, PARTICIPANTS).round(2),
    'Energy_Level': np.random.randint(1, 11, PARTICIPANTS),
    'Satisfaction_Rating': np.random.randint(1, 6, PARTICIPANTS),
    'Day': np.random.randint(1, DAYS + 1, PARTICIPANTS),
    'Feedback': [f'Participant feedback {i}' for i in range(1, PARTICIPANTS + 1)]
}

# Create DataFrame
df = pd.DataFrame(data)

# Display first few rows and basic information
print("Dataset Shape:", df.shape)
display(df.head())
display(df.describe())
```

```python
df.to_csv('Chrispo.csv', index=False)
```
